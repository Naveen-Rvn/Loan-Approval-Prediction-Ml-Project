```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("LoanApprovalPrediction.csv")
```

```python
data.head(5)
```

```python
obj = (data.dtypes == 'object')
print("Categorical variables:",len(list(obj[obj].index)))
```

```python
# Dropping Loan_ID column
data.drop(['Loan_ID'],axis=1,inplace=True)
```

```python
obj = (data.dtypes == 'object')
object_cols = list(obj[obj].index)

plt.figure(figsize=(18,36))
index = 1

for col in object_cols:
    y = data[col].value_counts()
    plt.subplot(11, 4, index)
    plt.xticks(rotation=90)
    sns.barplot(x=y.index, y=y.values)
    plt.title(col)  # Adding a title to each subplot
    index += 1

plt.tight_layout()  # Adjust subplots to fit into the figure area.
plt.show()
```

```python
from sklearn import preprocessing #model training

# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Identify object type columns
obj = (data.dtypes == 'object')

# Loop through the object columns and apply label encoding
for col in list(obj[obj].index):
    data[col] = label_encoder.fit_transform(data[col])
```

```python
# To find the number of columns with
# datatype==object
obj = (data.dtypes == 'object')
print("Categorical variables:",len(list(obj[obj].index)))
```

```python
plt.figure(figsize=(12,6))

sns.heatmap(data.corr(),cmap='BrBG',fmt='.2f',
			linewidths=2,annot=True)
```

```python
sns.catplot(x="Gender", y="Married",
			hue="Loan_Status",
			kind="bar",
			data=data)
```

```python
for col in data.columns:
    if data[col].dtype in ['float64', 'int64']:  # Check if the column is numeric
        data[col] = data[col].fillna(data[col].mean())
    else:
        data[col] = data[col].fillna(data[col].mode()[0]) # Fill non-numeric columns with the mode (most frequent value)

# Check for remaining NaN values
print(data.isna().sum())

```

```python
from sklearn.model_selection import train_test_split #testing the model we import train_test_split

X = data.drop(['Loan_Status'],axis=1) #Preparing the Data:
Y = data['Loan_Status']
X.shape,Y.shape #Checking the Shape of the Data:

X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
													test_size=0.4,
													random_state=1) #Splitting the Data into Training and Testing Sets:
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier # imports the KNeighborsClassifier class from the sklearn.neighbors
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC # imports the SVC class from the sklearn.svm module.
from sklearn.linear_model import LogisticRegression #imports the LogisticRegression class from the sklearn.linear_model module.

from sklearn import metrics

knn = KNeighborsClassifier(n_neighbors=3) #Initializing Classifiers:
rfc = RandomForestClassifier(n_estimators = 7,
							criterion = 'entropy',
							random_state =7)
svc = SVC()
lc = LogisticRegression()

# making predictions on the training set
for clf in (rfc, knn, svc,lc):
	clf.fit(X_train, Y_train)
	Y_pred = clf.predict(X_train)
	print("Accuracy score of ",
		clf.__class__.__name__,
		"=",100*metrics.accuracy_score(Y_train,
										Y_pred))
```

```python
# making predictions on the testing set
for clf in (rfc, knn, svc,lc):
	clf.fit(X_train, Y_train)
	Y_pred = clf.predict(X_test)
	print("Accuracy score of ",
		clf.__class__.__name__,"=",
		100*metrics.accuracy_score(Y_test,
									Y_pred))
```
